```python
browser_1 = webdriver.Chrome(executable_path=path)

"""Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
"""
# Load webpage:

browser_1.get('https://www.naukri.com/')

browser_1.maximize_window()

search_job = browser_1.find_element(By.CLASS_NAME,'suggestor-input')
search_job.send_keys('Data Analyst')

search_loc = browser_1.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
search_loc.send_keys('Banglore')

search_button = browser_1.find_element(By.CLASS_NAME,'qsbSubmit')
search_button.click()

# extracting jobs: 

job_ti = []
ti_tags = browser_1.find_elements(By.XPATH,"//a[@class='title fw500 ellipsis']")
for i in ti_tags:
    job_ti.append(i.text)

com_names = []
com_tags = browser_1.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']")
for i in com_tags:
    com_names.append(i.text)
com_names

exp_req = []
exp_tags = browser_1.find_elements(By.XPATH,"//span[@class='ellipsis fleft fs12 lh16 expwdth']")
for i in exp_tags:
    exp_req.append(i.text)


jobs_dt = pd.DataFrame()
jobs_dt['Job Titles']=job_ti
jobs_dt['Company names']=com_names
jobs_dt['Experience Required']=exp_req

jobs_dt.head(10)
```

```python
"""Q6: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the
search field.
You have to scrape 4 attributes of each sneaker:
1. Brand
2. Product Description
3. Price
As shown in the below image, you have to scrape the tick marked attributes."""

browser_6 = webdriver.Chrome(executable_path=path)

browser_6.get('https://www.flipkart.com/')

browser_6.maximize_window()

close_login6 = browser_6.find_element(By.XPATH,"/html/body/div[2]/div/div/button")
close_login6.click()

enter_item6 = browser_6.find_element(By.CLASS_NAME,'_3704LK')
enter_item6.send_keys('sneakers')

click_search6 = browser_6.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
click_search6.click()

time.sleep(3)

# Data extraction:

sneakers_brands = []
sneakers_desc = []
sneakers_price = []

for page in range(0,3):
    sneakers_title = browser_6.find_elements(By.XPATH,"//div[@class='_2WkVRV']")
    for j in sneakers_title:
        sneakers_brands.append(j.text)
    sneakers_da = browser_6.find_elements(By.XPATH,"//a[@class='IRpwTa']")
    sneakers_db = browser_6.find_elements(By.XPATH, "//a[@class='IRpwTa _2-ICcC']")
    for ka in sneakers_da:
        sneakers_desc.append(ka.text)
    for kb in sneakers_db:
        sneakers_desc.append(kb.text)
    sneakers_p = browser_6.find_elements(By.XPATH,"//div[@class='_30jeq3']")
    for l in sneakers_p:
        sneakers_price.append(l.text)
    next_button6 = browser_6.find_element(By.CLASS_NAME,'_1LKTO3')
    next_button6.click()
    time.sleep(3)
    

df_sneakers = pd.DataFrame()
df_sneakers['Brand Name']=sneakers_brands
df_sneakers['Description']=sneakers_desc
df_sneakers['Price']=sneakers_price

df_sneakers.head(100)
```

```python
"""Q9: Write a python program to scrape data for Top 1000 Quotes of All Time.
 The above task will be done in following steps:
 1.First get the webpage https://www.azquotes.com/
 2. Click on Top Quotes
 3. Than scrap a) Quote b) Author c) Type Of Quote
"""


browser_9 = webdriver.Chrome(executable_path=path)

browser_9.get('https://www.azquotes.com/')

browser_9.maximize_window()

click_topquotes_button = browser_9.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
click_topquotes_button.click()

top_quotes = []
authors = []
quote_types = []

for i in range(0,11):

    quote = browser_9.find_elements(By.XPATH,"//a[@class='title']")
    for i in quote:
        top_quotes.append(i.text)
    
    auth = browser_9.find_elements(By.XPATH,"//div[@class='author']")
    for j in auth:
        authors.append(j.text)
        
    type = browser_9.find_elements(By.XPATH,"//div[@class='tags']")
    for k in type:
        quote_types.append(k.text)

    time.sleep(3)

    next_page_button = browser_9.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    next_page_button.click()

df_quotes = pd.DataFrame()
df_quotes['Quote']=top_quotes
df_quotes['Authors']=authors
df_quotes['Type']=quote_types

df_quotes.head(1000)
```

```python
"""Q10: Write s python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, 
Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpage https://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make the DataFrame."""

browser_10 = webdriver.Chrome(executable_path=path)

browser_10.get('https://www.jagranjosh.com/')

browser_10.maximize_window()

click_gk_button10 = browser_10.find_element(By.XPATH,'/html/body/div/div[1]/div/div[1]/div/div[6]/div/div[1]/header/div[3]/ul/li[9]/a')
click_gk_button10.click()

search_pmlist = browser_10.find_element(By.XPATH,'/html/body/div/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a')
search_pmlist.click()

# Data extraction:

all_data = []
data = browser_10.find_elements(By.XPATH,'//td[@style]')
for i in data:
    all_data.append(i.text)
    
for i in range(4):
    all_data.pop()
    
all_data

sno = []
for i in range(0,86,5):
    sno.append(all_data[i])
    
pm_names = []
for i in range(1,87,5):
    pm_names.append(all_data[i])
    
bd_list = []
for i in range(2,88,5):
    bd_list.append(all_data[i])
    
term_list = []
for i in range(3,89,5):
    term_list.append(all_data[i].split('\n'))
    
remark_list = []
for i in range(4,90,5):
    remark_list.append(all_data[i])
    
df_pm = pd.DataFrame()
df_pm['Sno']=sno
df_pm['Prime Minister Names']=pm_names
df_pm['Born-Dead']=bd_list
df_pm['Term_list']=term_list
df_pm['Remarks']=remark_list


df_pm
```
